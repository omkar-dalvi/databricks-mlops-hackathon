# Ingest Data from ADLS to Databricks 

abfss://data@dynamicpricingdevadls.dfs.core.windows.net/competitorpricingdata.csv

```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

key_name = "adls_account_key"
scope_name = "adls"
account_key = ""

w.secrets.create_scope(scope=scope_name)
w.secrets.put_secret(scope=scope_name, key=key_name, string_value=account_key)

```

```python
storage_account = "dynamicpricingdevadls"

spark.conf.set(
    "fs.azure.account.key.dynamicpricingdevadls.dfs.core.windows.net",
    dbutils.secrets.get(scope="adls", key="adls_account_key"))

spark.conf.set(f"fs.azure.account.key.{storage_account}.blob.core.windows.net", dbutils.secrets.get(scope="adls", key="adls_account_key"))

```

```python
container = "data"


competitor_pricing_data_path = f"wasbs://{container}@{storage_account}.blob.core.windows.net/competitorpricingdata.csv"
daily_customer_behaviour_data_path = f"wasbs://{container}@{storage_account}.blob.core.windows.net/dailycustomerbehavior.csv"
tide_sales_data_path = f"wasbs://{container}@{storage_account}.blob.core.windows.net/tidesalesdata.csv"
tide_warehouse_inventory_data_path = f"wasbs://{container}@{storage_account}.blob.core.windows.net/tidewarehouseinventorysim.csv"



# Read data from ADLS into a Spark DataFrame (assuming CSV format)
competitor_pricing_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(competitor_pricing_data_path)

daily_customer_behaviour_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(daily_customer_behaviour_data_path)

tide_sales_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(tide_sales_data_path)

tide_warehouse_inventory_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(tide_warehouse_inventory_data_path)
```
