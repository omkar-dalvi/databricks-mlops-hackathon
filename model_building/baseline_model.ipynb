```python
import pandas as pd  
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_absolute_error, mean_squared_error  
from sklearn.preprocessing import StandardScaler  
```

```python
competitor_df = pd.read_csv("../data/competitorpricingdata.csv")
customer_behavior_df = pd.read_csv("../data/dailycustomerbehavior.csv")
sales_df = pd.read_csv("../data/tidesalesdata.csv")
inventory_df = pd.read_csv("../data/tidewarehouseinventorysim.csv")
```

```python
# Convert date columns to datetime  
# competitor_df['Date'] = pd.to_datetime(competitor_df['Date'])  
# customer_behavior_df['Date'] = pd.to_datetime(customer_behavior_df['Date'])  
# sales_df['TransactionDate'] = pd.to_datetime(sales_df['TransactionDate'])  
# inventory_df['Date'] = pd.to_datetime(inventory_df['Date'])  
  
# Merge datasets on 'Date'  
combined_data = sales_df.merge(competitor_df, left_on='TransactionDate', right_on='Date', how='left').merge(customer_behavior_df, left_on='TransactionDate', right_on='Date', how='left').merge(inventory_df, left_on='TransactionDate', right_on='Date', how='left')  
print(combined_data)
# Drop redundant 'Date' columns  
combined_data.drop(columns=['Date_x', 'Date_y', 'Date'], inplace=True)  
  
# Handle missing values (simple strategy: fill with mean)  
combined_data.fillna(combined_data.mean(), inplace=True)  
  
# Feature engineering: Calculate discount amount  
combined_data['DiscountAmount'] = combined_data['MRP_x'] - combined_data['FinalPrice']  
  
# Select features and target variable  
features = combined_data.drop(columns=['TransactionDate', 'SellingPrice'])  
target = combined_data['SellingPrice']  
  
# Standardize features  
scaler = StandardScaler()  
features_scaled = scaler.fit_transform(features)  
  
# Split data into training and testing sets  
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)  
  
# Train baseline model (Linear Regression)  
model = LinearRegression()  
model.fit(X_train, y_train)  
  
# Make predictions  
y_pred = model.predict(X_test)  
  
# Evaluate model performance  
mae = mean_absolute_error(y_test, y_pred)  
rmse = mean_squared_error(y_test, y_pred, squared=False)  
  
print(f'Mean Absolute Error: {mae}')  
print(f'Root Mean Squared Error: {rmse}')
```
